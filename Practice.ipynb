```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
# Windows users
chromedriver_path = 'C:/Users/renat/chromedriver_win32/chromedriver.exe'
executable_path = {'executable_path': chromedriver_path}
browser = Browser('chrome', **executable_path, headless=False) #headless=False means that all of the browser's actions will be displayed in a Chrome window so we can see them.
```

```python
#This code tells Splinter which site we want to visit by assigning the link to a URL

# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
#After executing the cell above, we will use BeautifulSoup to parse the HTML.

# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
title = html_soup.find('h2').text
title
```

We used our html_soup object we created earlier and chained find() to it to search for the <  h2 / > tag.

We've also extracted only the text within the HTML tags by adding .text to the end of the code.

```python
#Create a new variable tag_box, which will be used to store the results of a search. 
#In this case, we're looking for <div /> elements with a class of tags-box, 
#and we're searching for it in the HTML we parsed earlier and stored in the html_soup variable

# Scrape the top ten tags
tag_box = html_soup.find('div', class_='tags-box')

#"tags" variable will hold the results of a find_all, 
#but this time we're searching through the parsed results stored in our tag_box variable to 
#find <a /> elements with a tag class.

# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
#Scrape Across Pages

url = 'http://quotes.toscrape.com/'
browser.visit(url)

#A for loop with five iterations (pages 1-6)
for x in range(1, 6):
    
    #Create an HTML object, assigned to the html variable
    html = browser.html
    
    #Use BeautifulSoup to parse the html object
    quote_soup = soup (html, 'html.parser')
    
    #Use BeautifulSoup to find all <span /> tags with a class of "text"
    quotes = quote_soup.find_all('span', class_='text')
    
    #Print statements wrapped in another for loop that will print each quote parsed by BeautifulSoup
    for quote in quotes:
        print('page:', x, '----------')
        print(quote.text)
    
    #Use Splinter to click the "Next" button
    browser.links.find_by_partial_text('Next')
```

```python
#Scrape the book URL list on the first page of http://books.toscrape.com/

url = 'http://books.toscrape.com/'
browser.visit(url)

#Create an HTML object, assigned to the html variable
html = browser.html
    
#Use BeautifulSoup to parse the html object
url_soup = soup (html, 'html.parser')
    
#Use BeautifulSoup to find all <ul> tags with a class of "nav nav-list"
urls = url_soup.find_all('ul', class_='nav nav-list')

#Print each url parsed by BeautifulSoup
for url in urls:
    print(url.text)
```
